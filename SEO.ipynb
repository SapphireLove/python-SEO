Agreggate daily search console data into weekly starting Monday and ending on Sunday.

```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('VIS-Q1.csv')

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract week number and year
df['Week_Number'] = df['Date'].dt.isocalendar().week
df['Year'] = df['Date'].dt.year

# Aggregate data by week and year
weekly_agg = df.groupby(['Year', 'Week_Number'])[['Clicks', 'Impressions']].sum().reset_index()

# Function to get start and end dates of a week
def get_week_dates(year, week_num):
    start_date = pd.to_datetime(f'{year}-{week_num}-1', format='%Y-%W-%w')
    end_date = start_date + pd.DateOffset(days=6)
    return f"{start_date.strftime('%m/%d/%Y')} - {end_date.strftime('%m/%d/%Y')}"

# Create 'Week' column in the desired format
weekly_agg['Week'] = weekly_agg.apply(lambda row: get_week_dates(row['Year'], row['Week_Number']), axis=1)

# Select and reorder columns
weekly_agg = weekly_agg[['Week', 'Clicks', 'Impressions']]

# Save to CSV
weekly_agg.to_csv('weekly_aggregated_data_vis_q1.csv', index=False)

# Display the first 5 rows
print(weekly_agg.head().to_markdown(index=False, numalign='left', stralign='left'))
```

Extract All Meta Titles and Desc from URLs in List

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time

# Load your data from the CSV file
df = pd.read_csv('All WWW URLs.csv')

# Function to extract meta tags
def extract_meta(url):
    try:
        response = requests.get(url, timeout=10)  # Timeout for safety
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')
        title = soup.find('title').text if soup.find('title') else ''
        description = soup.find('meta', attrs={'name': 'description'})
        description = description['content'] if description else ''
        return title, description
    except (requests.exceptions.RequestException, AttributeError) as e:
        print(f"Error processing {url}: {e}")
        return '', ''  # Return empty values in case of errors

# Iterate through URLs and extract meta tags
meta_data = []
for url in df['Unique WWW URLs']:
    title, description = extract_meta(url)
    meta_data.append([url, title, description])
    time.sleep(1)  # Polite delay

# Create DataFrame and save to CSV
meta_df = pd.DataFrame(meta_data, columns=['Unique WWW URLs', 'Meta Title', 'Meta Description'])
meta_df.to_csv('all_meta_titles_and_desc.csv', index=False)

print("Extraction complete. Results saved to all_meta_titles_and_desc.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import csv
from urllib.parse import urljoin

def crawl_and_extract_urls(base_url, visited_urls=None):
    """Crawls a website and extracts internal URLs.

    Args:
        base_url (str): The starting URL for the crawl.
        visited_urls (set, optional): A set to keep track of visited URLs. Defaults to None.
    """

    if visited_urls is None:
        visited_urls = set()

    response = requests.get(base_url)
    response.raise_for_status()

    soup = BeautifulSoup(response.content, 'html.parser')

    extracted_urls = []
    for link in soup.find_all('a'):
        href = link.get('href')
        if href and href.startswith('/') or urljoin(base_url, href).startswith(base_url):
            absolute_url = urljoin(base_url, href)

            if absolute_url not in visited_urls:
                extracted_urls.append(absolute_url)
                visited_urls.add(absolute_url)
                crawl_and_extract_urls(absolute_url, visited_urls)

    return extracted_urls

# Start crawling
base_url = "https://www.safelinkwireless.com"
all_urls = crawl_and_extract_urls(base_url)

# Save to CSV
with open('safelink_all.csv', 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    for url in all_urls:
        csv_writer.writerow([url])
```

```python
import requests
import xml.etree.ElementTree as ET
import os

def fetch_and_save_sitemap_urls(sitemap_url, output_file_path):
    try:
        # Step 1: Fetch the sitemap
        response = requests.get(sitemap_url)
        response.raise_for_status()  # Raises an HTTPError if the response status code is 4XX/5XX

        # Step 2: Parse the XML to find all URLs
        root = ET.fromstring(response.content)
        urls = []

        # Namespace to parse the XML properly
        namespace = {'ns': 'http://www.sitemaps.org/schemas/sitemap/0.9'}

        for url in root.findall('ns:url/ns:loc', namespace):
            urls.append(url.text)

        # Ensure the output directory exists
        output_directory = os.path.dirname(output_file_path)
        if not os.path.exists(output_directory):
            os.makedirs(output_directory)

        # Step 3: Write URLs to the specified file path
        with open(output_file_path, 'w') as file:
            for url in urls:
                file.write(url + '\n')

        print(f"File created successfully with {len(urls)} URLs at {output_file_path}")

    except requests.HTTPError as http_err:
        print(f"HTTP error occurred: {http_err}")
    except Exception as err:
        print(f"An error occurred: {err}")

# Specify the sitemap URL and the output file path
sitemap_url = "https://www.healthcoachpempi.com/sitemap.xml"
output_file_path = "/content/untitled"  # File path updated as requested

# Call the function with the specified parameters
fetch_and_save_sitemap_urls(sitemap_url, output_file_path)
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
import csv

def is_external(url, main_domain):
    """Check if a URL is external."""
    return urlparse(url).netloc and urlparse(url).netloc != main_domain

def is_internal(url, main_domain):
    """Check if a URL is internal."""
    return urlparse(url).netloc == main_domain

def get_links(url, main_domain):
    """Get all unique internal and external links from a URL."""
    internal_links = set()
    external_links = set()
    try:
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            for link_tag in soup.find_all('a', href=True):
                full_link = urljoin(url, link_tag['href'])
                if is_internal(full_link, main_domain):
                    internal_links.add(full_link)
                elif is_external(full_link, main_domain):
                    external_links.add((full_link, url))
    except requests.exceptions.RequestException as e:
        print(f"Error fetching {url}: {e}")
    return internal_links, external_links

# Main URL
main_url = 'https://www.straighttalk.com'
main_domain = urlparse(main_url).netloc

# Googlebot user agent string
headers = {
    'User-Agent': 'Mozilla/5.0 (compatible; Googlebot/2.1; +http://www.google.com/bot.html)'
}

# Sets to keep track of visited and to-be-visited URLs
visited_urls = set()
urls_to_visit = {main_url}
external_links_found = set()

# Start crawling
while urls_to_visit:
    current_url = urls_to_visit.pop()
    print(f"Crawling: {current_url}")
    visited_urls.add(current_url)

    internal_links, external_links = get_links(current_url, main_domain)
    external_links_found.update(external_links)

    # Add new internal links to the queue
    urls_to_visit.update(internal_links - visited_urls)

# Write the external links found to a CSV file
with open('external_links.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['External Link', 'Found On Page'])
    for link, source in external_links_found:
        writer.writerow([link, source])

print("External links have been saved to 'external_links.csv'")
```

```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('urls.csv')  # Replace 'urls.csv' with your actual file name

# Assuming the columns are named 'URLs_A' and 'URLs_B'
urls_a = set(df['URLs_A'])
urls_b = set(df['URLs_B'])

# Find URLs in B not in A
urls_in_b_not_in_a = urls_b - urls_a

# Output the result
print("URLs in B that are not in A:")
for url in urls_in_b_not_in_a:
    print(url)

# Save the unique URLs to a new CSV file
output_file = 'unique_urls.csv'
with open(output_file, 'w', newline='', encoding='utf-8') as file:
    pd.DataFrame(list(urls_in_b_not_in_a), columns=["Unique URLs in B"]).to_csv(file, index=False)

print(f"Unique URLs have been saved to {output_file}")
```
